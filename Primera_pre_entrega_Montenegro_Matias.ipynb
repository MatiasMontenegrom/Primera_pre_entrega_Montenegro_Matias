```python

# Seleccionar las columnas relacionadas con Sleep Duration
sleep_columns = [
    'Sleep Duration_Less than 5 hours',
    'Sleep Duration_7-8 hours',
    'Sleep Duration_More than 8 hours'
]

# Calcular la proporción de estudiantes en cada categoría
sleep_distribution = df[sleep_columns].sum() / len(df)

# Crear un gráfico de barras para la distribución
plt.figure(figsize=(8, 5))
sleep_distribution.plot(kind='bar', color=['red', 'green', 'blue'], alpha=0.8)
plt.title('Distribución de Sleep Duration', fontsize=14)
plt.ylabel('Proporción de estudiantes', fontsize=12)
plt.xlabel('Categoría de duración del sueño', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Seleccionar las columnas relacionadas con Sleep Duration y Depression
sleep_columns = [
    'Sleep Duration_Less than 5 hours',
    'Sleep Duration_7-8 hours',
    'Sleep Duration_More than 8 hours'
]
depression_column = 'Depression_Yes'

# Calcular la proporción de estudiantes con depresión en cada categoría de sueño
depression_rates = {}
for col in sleep_columns:
    depression_rates[col] = df[df[col] == 1][depression_column].mean()

# Convertir a DataFrame para facilitar la visualización
depression_rates_df = pd.DataFrame.from_dict(depression_rates, orient='index', columns=['Depression Rate'])

# Crear un gráfico de barras
plt.figure(figsize=(8, 5))
depression_rates_df['Depression Rate'].plot(kind='bar', color=['red', 'green', 'blue'], alpha=0.8)
plt.title('Tasas de depresión según duración del sueño', fontsize=14)
plt.ylabel('Proporción de estudiantes con depresión', fontsize=12)
plt.xlabel('Categoría de duración del sueño', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from scipy.stats import chi2_contingency


# Crear una tabla de contingencia
sleep_depression_table = pd.DataFrame({
    'Less than 5 hours': [
        df[(df['Sleep Duration_Less than 5 hours'] == 1) & (df['Depression_Yes'] == 1)].shape[0],
        df[(df['Sleep Duration_Less than 5 hours'] == 1) & (df['Depression_Yes'] == 0)].shape[0]
    ],
    '7-8 hours': [
        df[(df['Sleep Duration_7-8 hours'] == 1) & (df['Depression_Yes'] == 1)].shape[0],
        df[(df['Sleep Duration_7-8 hours'] == 1) & (df['Depression_Yes'] == 0)].shape[0]
    ],
    'More than 8 hours': [
        df[(df['Sleep Duration_More than 8 hours'] == 1) & (df['Depression_Yes'] == 1)].shape[0],
        df[(df['Sleep Duration_More than 8 hours'] == 1) & (df['Depression_Yes'] == 0)].shape[0]
    ]
}, index=['Depressed', 'Not Depressed'])

print("Tabla de contingencia:")
print(sleep_depression_table)

# Realizar el test de chi-cuadrado
chi2, p, dof, expected = chi2_contingency(sleep_depression_table)

print("\nResultados del test de Chi-cuadrado:")
print(f"Chi2: {chi2}")
print(f"P-valor: {p}")
print(f"Grados de libertad: {dof}")
print("Tabla esperada:")
print(pd.DataFrame(expected, index=sleep_depression_table.index, columns=sleep_depression_table.columns))
```

Dado que p-valor = 0.369 > 0.05, no podemos rechazar la hipótesis nula. Esto implica que:

    No hay evidencia estadísticamente significativa de que la duración del sueño esté asociada con la depresión en este conjunto de datos.
    Las diferencias observadas podrían deberse al azar.

```python


# Crear una nueva columna numérica para Sleep Duration
df['Sleep Duration Numeric'] = (
    df['Sleep Duration_Less than 5 hours'] * 1 +
    df['Sleep Duration_7-8 hours'] * 2 +
    df['Sleep Duration_More than 8 hours'] * 3
)

# Calcular la correlación de Pearson
correlation = df['Sleep Duration Numeric'].corr(df['Academic Pressure'])

print(f"Correlación entre Sleep Duration y Academic Pressure: {correlation:.2f}")
```

Cambios en la duración del sueño no parecen estar asociados con cambios en la presión académica.
No hay evidencia de que dormir más o menos influya directamente en el nivel de presión académica experimentado por los estudiantes.

```python
import statsmodels.api as sm

# Crear la variable numérica para Sleep Duration si aún no existe
df['Sleep Duration Numeric'] = (
    df['Sleep Duration_Less than 5 hours'] * 1 +
    df['Sleep Duration_7-8 hours'] * 2 +
    df['Sleep Duration_More than 8 hours'] * 3
)

# Seleccionar las variables independientes y dependiente
X = df[['Sleep Duration Numeric', 'Study Hours']]  # Variables independientes
y = df['Academic Pressure']  # Variable dependiente

# Agregar una constante para el término independiente
X = sm.add_constant(X)

# Ajustar el modelo de regresión lineal
model = sm.OLS(y, X).fit()

# Mostrar los resultados del modelo
print(model.summary())
```

El modelo de regresión muestra que la duración del sueño y las horas de estudio no tienen una relación significativa con la presión académica en este conjunto de datos.
R² bajo y p-valores altos indican que otras variables no incluidas en el modelo pueden estar influyendo en la presión académica o que la relación no es lineal.

```python


# Codificar Gender y Financial Stress si no está hecho
df['Gender_Male'] = df['Gender_Male'].astype(int)  # Asumiendo que 'Gender_Male' es 0/1
df['Financial Stress'] = df['Financial Stress'].astype(int)  # 0 para no, 1 para sí

# Crear la variable numérica de Sleep Duration
df['Sleep Duration Numeric'] = (
    df['Sleep Duration_Less than 5 hours'] * 1 +
    df['Sleep Duration_7-8 hours'] * 2 +
    df['Sleep Duration_More than 8 hours'] * 3
)

# Definir las variables independientes (con interacciones)
X = df[['Sleep Duration Numeric', 'Gender_Male', 'Financial Stress']]
X['Sleep_Gender_Interaction'] = X['Sleep Duration Numeric'] * X['Gender_Male']  # Interacción Sleep Duration * Gender
X['Sleep_Financial_Interaction'] = X['Sleep Duration Numeric'] * X['Financial Stress']  # Interacción Sleep Duration * Financial Stress

# Agregar una constante para el término independiente
X = sm.add_constant(X)

# Variable dependiente
y = df['Depression_Yes']  # Asumiendo que la variable de depresión es binaria (0/1)

# Ajustar el modelo de regresión
model = sm.Logit(y, X).fit()  # Usamos regresión logística ya que Depression es binaria

# Mostrar los resultados
print(model.summary())
```

Aunque el estrés financiero tiene un efecto significativo sobre la depresión, la duración del sueño y las interacciones entre la duración del sueño, el género y el estrés financiero no muestran una moderación significativa de la relación entre el sueño y la depresión.

Conclusiones Generales:

    Relación entre la duración del sueño y la depresión:
        No encontramos evidencia fuerte de una relación significativa entre la duración del sueño y la depresión. Los coeficientes de las regresiones y las pruebas de chi-cuadrado no indican que la duración del sueño tenga un impacto directo en la probabilidad de desarrollar depresión. Sin embargo, la duración del sueño podría ser un factor indirecto que afecta otros aspectos como el estrés financiero o el bienestar general, que a su vez influyen en la depresión.

    Género y su impacto en la depresión:
        El género no mostró un efecto significativo sobre la depresión en el análisis. La diferencia entre hombres y mujeres en cuanto a la probabilidad de tener depresión no fue estadísticamente significativa. Es posible que otros factores, como las expectativas sociales o los estilos de vida asociados con el género, puedan influir, pero estos no fueron capturados en el análisis actual.

    Impacto del estrés financiero:
        El estrés financiero tiene una influencia significativa en la depresión. Los resultados muestran que los estudiantes que experimentan altos niveles de estrés financiero tienen una mayor probabilidad de presentar síntomas de depresión. Este hallazgo es consistente con la literatura existente que sugiere que las preocupaciones económicas pueden tener efectos negativos en la salud mental.

    Moderación por género y estrés financiero:
        Las interacciones entre la duración del sueño y el género o entre la duración del sueño y el estrés financiero no fueron significativas. Esto indica que, en este contexto, el género y el estrés financiero no moderan significativamente la relación entre la duración del sueño y la depresión. Aunque el estrés financiero en sí tiene un impacto significativo sobre la depresión, no parece haber una interacción directa con la duración del sueño en este modelo.
